```python
import numpy as np

def bisection_method(func, a, b, tol=0.005):
    if func(a) * func(b) >= 0:
        raise ValueError("The function must have different signs at a and b.")

    while (b - a) / 2.0 > tol:
        c = (a + b) / 2
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

def f1(x):
    return x**2 - 2

def f2(x):
    return x**5 - 6*x**3 + 2*x - 2

def f3(x):
    return np.cos(x) - x

def f4(x):
    return 2 * np.sin(x) - np.sqrt(x + 1)


roots = {}
roots['f1'] = bisection_method(f1, 0, 4)


print("Roots:")
print(f"f1: {roots['f1']:.5f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt


def g(t):
    return t**2 - np.cos(t)**2

t_values = np.linspace(-np.pi, np.pi, 400)
g_values = g(t_values)

plt.figure(figsize=(10, 5))
plt.plot(t_values, g_values, label='g(t) = t^2 - cos^2(t)', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Graph of g(t) = t^2 - cos^2(t)')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.grid()
plt.legend()
plt.show()


num_solutions = np.sum(np.diff(np.sign(g_values)) != 0)
print(f"number to solutions of the equation t^2 = cos^2(t): {num_solutions}")


def bisection_method(func, a, b, tol=0.005):
    if func(a) * func(b) >= 0:
        raise ValueError("The function must have different signs at a and b.")

    while (b - a) / 2 > tol:
        c = (a + b) / 2.0
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2.0


solutions = []
for i in range(len(g_values) - 1):
    if g_values[i] * g_values[i + 1] < 0:
        root = bisection_method(g, t_values[i], t_values[i + 1])
        solutions.append(root)


print("estimate solutions t^2 = cos^2(t):")
for sol in solutions:
    print(f"{sol:.5f}")
```

##PUNTO 2.5.3

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones de demanda y oferta
def D(p):
    return 50000 * p + 10000 * p**2

def S(p):
    return (1/3) * p**2 + 2 * p

# (a) Graficar las funciones de demanda y oferta
p_values = np.linspace(0, 10, 400)
D_values = D(p_values)
S_values = S(p_values)

plt.figure(figsize=(10, 6))
plt.plot(p_values, D_values, label='Demanda D(p)', color='blue')
plt.plot(p_values, S_values, label='Oferta S(p)', color='orange')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Grficas de Demanda y Oferta')
plt.xlabel('Precio (p)')
plt.ylabel('Unidades')
plt.legend()
plt.grid()
plt.show()


for price in np.linspace(0, 10, 11):
    print(f"Precio: {price:.2f}, D(p): {D(price):.2f}, S(p): {S(price):.2f}")


def equilibrium_price(func_d, func_s, a, b, tol=0.005):
    def f(p):
        return func_d(p) - func_s(p)

    f_a = f(a)
    f_b = f(b)

    print(f"f({a}) = {f_a:.2f}, f({b}) = {f_b:.2f}")

    if f_a * f_b >= 0:
        raise ValueError("La funcin debe tener signos diferentes en a y b.")

    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f(c) == 0:
            return c
        elif f_a * f(c) < 0:
            b = c
            f_b = f(c)
        else:
            a = c
            f_a = f(c)
    return (a + b) / 2.0


price_equilibrium = equilibrium_price(D, S, 0, 10)
print(f"Precio de equilibrio: {price_equilibrium:.5f}")


units_at_equilibrium = D(price_equilibrium)
print(f"Unidades producidas a precio de equilibrio: {units_at_equilibrium:.5f}")


if price_equilibrium < 10:
    print("Si los productores suben el precio por encima del precio de equilibrio, habr un excedente.")
else:
    print("Si los productores bajan el precio por debajo del precio de equilibrio, habr una escasez.")


print("Si los productores aumentan la produccin, pueden satisfacer la demanda creciente.")
print("Si los productores disminuyen la produccin, pueden enfrentar escasez si la demanda se mantiene constante.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar


def area(x):
    return - (500 * x - 2 * x**2)


result = minimize_scalar(area, bounds=(0, 250), method='bounded')


optimal_x = result.x
optimal_y = 500 - 2 * optimal_x

print(f"Dimensiones ptimas del campo:")
print(f"Lado perpendicular al ro (x): {optimal_x:.2f} yards")
print(f"Lado paralelo al ro (y): {optimal_y:.2f} yards")


x_values = np.linspace(0, 250, 100)
area_values = 500 * x_values - 2 * x_values**2

plt.figure(figsize=(10, 6))
plt.plot(x_values, area_values, label='rea A(x)', color='blue')
plt.axvline(optimal_x, color='red', linestyle='--', label='x ptimo')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Maximizacin del rea del Campo')
plt.xlabel('Longitud de los lados perpendiculares al ro (x)')
plt.ylabel('rea (A)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar


def profit(p):
    return -(-50 * p**2 + 1000 * p - 3750)

result = minimize_scalar(profit, bounds=(0, 20), method='bounded')


optimal_price = result.x


max_profit = -result.fun

print(f"Precio ptimo para maximizar el beneficio: ${optimal_price:.2f}")
print(f"Beneficio mximo: ${max_profit:.2f}")

p_values = np.linspace(0, 20, 100)
profit_values = -(-50 * p_values**2 + 1000 * p_values - 3750)

plt.figure(figsize=(10, 6))
plt.plot(p_values, profit_values, label='Beneficio P(p)', color='blue')
plt.axvline(optimal_price, color='red', linestyle='--', label='Precio ptimo')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Maximizacin del Beneficio del Alfarero')
plt.xlabel('Precio por pieza (p)')
plt.ylabel('Beneficio (P)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def h_function(t):
    return t**4 - 1

t_values = np.linspace(-1, 1.999, 400)
h_values = h_function(t_values)

plt.figure(figsize=(8, 6))
plt.plot(t_values, h_values, label=r'$h(t) = t^4 - 1$', color='b')

plt.scatter(-1, h_function(-1), color='red', zorder=5, label="t = -1 (0, mnimo local)")
plt.scatter(0, h_function(0), color='green', zorder=5, label="t = 0 (-1, mnimo global)")

plt.axvline(x=2, color='orange', linestyle='--', label="Lmite en t=2")

plt.title('Grfica de $h(t) = t^4 - 1$ en el intervalo [-1, 2)')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def heaviside(t):
    return np.where(t < 0, 0, 1)

t_values = np.linspace(-1, 1, 400)
h_values = heaviside(t_values)

h_min = np.min(h_values)
h_max = np.max(h_values)

plt.figure(figsize=(8, 6))
plt.plot(t_values, h_values, label=r'$H(t)$ (Funcin Heaviside)', color='b')

plt.scatter([-1, 0, 1], [heaviside(-1), heaviside(0), heaviside(1)], color='red', zorder=5, label="Puntos clave")

plt.title('Grfica de la funcin Heaviside en el intervalo [-1, 1]')
plt.xlabel('t')
plt.ylabel('H(t)')
plt.legend()
plt.grid(True)
plt.show()

{
    "H(-1)": heaviside(-1),
    "H(1)": heaviside(1),
    "Mximo en [-1, 1]": h_max,
    "Mnimo en [-1, 1]": h_min,
    "Existe c tal que H(c) = 0.5?": "No"
}
```

```python
import numpy as np
import matplotlib.pyplot as plt

def g_function(t):
    return np.where(t == 0, 1, np.abs(t))

t_values = np.linspace(-1, 1, 400)
g_values = g_function(t_values)

g_min = np.min(g_values)
g_max = np.max(g_values)

plt.figure(figsize=(8, 6))
plt.plot(t_values, g_values, label=r'$g(t)$', color='b')

plt.scatter(0, g_function(0), color='red', zorder=5, label="t = 0 (g(0) = 1)")
plt.scatter([-1, 1], [g_function(-1), g_function(1)], color='green', zorder=5, label="t = -1, 1 (|t|)")

plt.title('Grfica de $g(t)$ en el intervalo [-1, 1]')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.legend()
plt.grid(True)
plt.show()

{
    "g(0)": g_function(0),
    "g(-1)": g_function(-1),
    "g(1)": g_function(1),
    "Mximo en [-1, 1]": g_max,
    "Mnimo en [-1, 1]": g_min
}
```

# PUNTO 2.5.9

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.sin(np.pi * t) + t

def g(t):
    return t**2

t_values = np.linspace(0, 1, 400)
f_values = f(t_values)
g_values = g(t_values)

h_values = f_values - g_values

plt.figure(figsize=(8, 6))
plt.plot(t_values, f_values, label=r'$f(t)$', color='blue')
plt.plot(t_values, g_values, label=r'$g(t)$', color='green')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

c_estimated = t_values[np.argmin(np.abs(h_values))]
plt.scatter(c_estimated, f(c_estimated), color='red', label=f'Punto c  {c_estimated:.3f} donde f(c) = g(c)')

plt.title('Demostracin grfica de $f(c) = g(c)$ en [0, 1]')
plt.xlabel('t')
plt.ylabel('Valor de las funciones')
plt.legend()
plt.grid(True)
plt.show()

{
    "Punto aproximado donde f(c) = g(c)": c_estimated
}
```
